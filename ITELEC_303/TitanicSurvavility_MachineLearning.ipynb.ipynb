<h1> TITANIC SURVIVAVILITY </h1>

```python
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Titanic.csv') 
```

```python
df
```

```python
# Drop irrelevant columns
df = df[["Survived", "Pclass", "Sex", "Age", "Embarked"]]

# Drop missing values
df = df.dropna()

# Encode categorical variable
df["Sex"] = df["Sex"].map({"male": 0, "female": 1})
df["Embarked"] = df["Embarked"].map({"C":0, "Q":1, "S":2})
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X = df[["Pclass", "Sex", "Age","Embarked"]]
y = df["Survived"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model = RandomForestClassifier()
model.fit(X_train, y_train)
# model.fit(X,y)
```

<h1> Accuracy Test </h1>

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

<h3>Accuracy Result</h3>
<hr>
<h4> When the result of the accuracy is above 0.6 it means the model is picking up patterns it is now <strong> LEARNING </strong></h4>

<h1> TITANIC SURVAVILITY USING LOGISTIC REGRESSION </h1>

```python
df2 = pd.read_csv("Titanic.csv")
df2
```

```python
# Keep only useful columns
df2 = df2[["Survived", "Pclass", "Sex", "Age","Fare" ,"Embarked"]]

# Drop missing values
df2 = df2.dropna()
```

```python
df2
```

```python
# Encode 'Sex' and 'Embarked'
df2["Sex"] = df2["Sex"].map({"male": 0, "female": 1})
df2["Embarked"] = df2["Embarked"].map({"S": 0, "C": 1, "Q": 2})
```

```python
df2
```

<h1>selecting X and y </h1>

```python
X = df2[["Pclass", "Sex", "Age", "Fare", "Embarked"]]
y = df2["Survived"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

<h1> Manual Testing </h1>

```python
# [Pclass=3, Sex=0 (male), Age=22, Fare=7.25, Embarked=0 (S)]
model.predict([[3, 0, 22, 7.25, 0]])
```

```python
kid = model.predict([[3, 0, 10, 7.00, 0]])
print("Kid prediction:", kid)
adult_male = model.predict([[2, 0, 30, 13.00, 2]])
print("Adult male prediction:", adult_male)
adult_female = model.predict([[1, 1, 25, 80.00, 1]])
print("Adult female prediction:", adult_female)
```

```python
def check(pred):
    return "Survived" if pred[0] == 1 else "Did NOT survive"

print("Kid:", check(kid))
print("Adult Male:", check(adult_male))
print("Adult Female:", check(adult_female))
```

<h4> Passenger 1 ( young kid age 1-15 ) </h4> <b>INSIGHT</b>-Although the passenger is a child, the model predicts they would not survive, likely because many 3rd-class children had low survival rates unless accompanied by adults.

<h4> Passenger 2 ( adult male )</h4><b>INSIGHT</b> -The model predicts the adult male would not survive, which is consistent with historical trends showing that men especially those in lower classes had the lowest survival chances.

<h4> Passenger 3 ( adult female )</h4><b>INSIGHT</b> -The model predicts the adult female survived, aligning with the “women and children first” evacuation practice during the Titanic tragedy.
